#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# TODO
#Using the mean normalization method for each population and the creating a DataFrame to work with the normalized prices for each population
norm_prices = pd.DataFrame()

for i in sb['level5'].unique():
    aux = (sb[sb['level5'] == i]['price'] - sb[sb['level5'] == i]['price'].mean()) / sb[sb['level5'] == i]['price'].std()
    norm_prices[i] = [aux]

#Plot for the whole four populations
for i in norm_prices:
    plt.hist(norm_prices[i], label='Norm Prices ' + i, bins=14, alpha=.4)

plt.legend()
plt.show()

#Subplots for each population
fig, axs= plt.subplots(2, 2, figsize = (10,10))
aux_x = 0 
aux_y = 0
for i in norm_prices:
    axs[aux_x, aux_y].hist(norm_prices[i])
    axs[aux_x, aux_y].set_title(i)
    aux_y += 1

    if aux_y > 1:
        aux_x += 1
        aux_y = 0

plt.show()
```
